```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import glob
```

### Uber speed data

```python
path = 'uber_data'
all_files = glob.glob(path + "/*.csv")

list = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, parse_dates=['utc_timestamp'])
    list.append(df)

speed_data = pd.concat(list, axis=0, ignore_index=True)
speed_data.to_csv( "speed_data.csv", index=False, encoding='utf-8-sig')
speed_data.tail(20)
```
